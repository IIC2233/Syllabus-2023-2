# DCCorredor de Propiedades

```python
#Añadir solución
from abc import ABC, abstractmethod


class Inmueble(ABC):
    
    def __init__(self,n,m,banios, habs) -> None:
        self.numero = n
        self.m2= m
        self.banios = banios
        self.habitaciones = habs

    @property 
    @abstractmethod  
    def precio(self):
        pass

    @abstractmethod
    def describir(self):
        pass


class Casa(Inmueble):
    
    def __init__(self,n,m,banios, habs,par,niv) -> None:
        super().__init__(n,m,banios, habs)
        self.niveles = niv
        self.par = par

    @property
    def precio(self):
        return 1000*self.banios* self.habitaciones*self.niveles* self.m2/ \
    (10+10*int(self.par))

    def describir(self):
        print(f"""Casa N°{self.numero}
M2: {self.m2}
Baños: {self.banios}
Piezas: {self.habitaciones}
Pisos: {self.niveles}
Pariada: {self.par}
------------------------
Precio: {self.precio}
""")


class Depto(Inmueble):


    def __init__(self,n,m,banios, habs,elev,piso) -> None:
        super().__init__(n,m,banios, habs)
        self.elevador = elev
        self.piso = piso


    @property
    def precio(self):
        return 1000*self.banios* self.habitaciones* self.m2/ \
                            (20-10*int(self.elevador))*self.piso
    
    
    def describir(self):
        print(
            f"""Depto N°{self.numero} Ubicado en el Piso {self.piso}
            M2: {self.m2}
            Baños: {self.banios}
            Piezas: {self.habitaciones}
            Piso: {self.piso}
            Elevador: {self.elevador}
            ------------------------
            Precio: {self.precio}
            """
        )
    
casa = Casa(1, 40, 3, 2, False, 2)
casa.describir()
casa.precio
depto = Depto(1, 20, 2, 3, True, 10)
depto.describir()
```

## DC(A)Cademia

```python
from abc import ABC, abstractmethod


class EstudianteUc(ABC): 
    #Clase base

    def __init__(self, nombre, numero_estudiante, semestre_actual):
        self.nombre = nombre
        self.numero_estudiante = numero_estudiante
        self.semestre_actual = semestre_actual

    def asistir_clase(self, clase):
        print(f"Asistiendo a la clase de {clase}")

    def procrastinar(self):
        print("Zzzzzz...")

    @abstractmethod
    def ir_a_la_universidad(self):
        pass

    @abstractmethod
    def pasar_de_largo(self):
        pass


class EstudianteDCC(EstudianteUc): 
    # Heredamos de EstudianteUC

    def __init__(self, sist_operativo, version_python, *args, **kwargs):
        super().__init__(*args, **kwargs) 
        #Heredamos los atributos necesarios (lo mismo para las otras clases que heredan)
        self.sist_operativo = sist_operativo
        self.version_python = version_python

    def programar(self):
        print(f"Programando en python versión {self.version_python}")

    def pasar_de_largo(self):
        print("No entiendo la multiherencia...")

    def ir_a_la_universidad(self):
        print("Yendo al DCC...")


class Seleccionado(EstudianteDCC): 
    # Heredamos de EstudianteDCC

    def __init__(self, seleccion, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.seleccion = seleccion
    
    def juego(self, contrincante):
        print(f"Jugando contra {contrincante}")

    def ir_a_la_universidad(self):
        print("Yendo a las canchas de deporte")


class SeleccionadoDCC(Seleccionado, EstudianteDCC): 
    # Hacemos multiherencia heredando tanto de seleccionado y de estudiante DCC

    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def seleccionado(self):
        print(f"Llegue al top, estoy compitiendo profesionalmente en la seleccion {self.seleccion}!")


Juan = SeleccionadoDCC(nombre="Juan",numero_estudiante=1232,
semestre_actual=2,sist_operativo="python",
version_python=3.10,seleccion="Chilena")
Juan.seleccionado()
```

```python
class Animal():

    def __init__(self, nombre, clasificacion):
        self.nombre = nombre
        self.clasificacion = clasificacion
    
    def exhibicion(self):
        print(f"El animal es un {self.nombre} y es un {self.clasificacion}")
```

```python
class Carnivoro(Animal):

    def __init__(self, carne_consumida, hambre, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.carne_consumida = carne_consumida
        self.hambre = hambre
        
    def exhibicion(self):
        super().exhibicion()
        print(f'Ha comido {self.carne_consumida}kg de carne y quedó con {self.hambre} hambre.')

class Herbivoro(Animal):

    def __init__(self, verduras_consumidas, *args,  **kwargs):
        super().__init__(*args, **kwargs)
        self.verduras_consumidas = verduras_consumidas
    
    def exhibicion(self):
        super().exhibicion()
        print(f'Ha comido {self.verduras_consumidas} verduras hoy.')
```

```python
class Omnivoro(Carnivoro, Herbivoro):

    def __init__(self, *args,  **kwargs):
        super().__init__(*args, **kwargs)

    def exhibicion(self):
        super().exhibicion() 
```

```python
Chanchito = Omnivoro(nombre="Chanchito", clasificacion="Mamífero", carne_consumida=2, 
                     hambre="mucha", verduras_consumidas=5)

Chanchito.exhibicion()
```

```python
from random import randint


class Lista2(list):

    def repetidos(self):
        aux = []
        repetidos = []
        for elemento in self:
            if elemento not in aux:
                aux.append(elemento)
            else:
                repetidos.append(elemento)
        for elemento in repetidos:
            self.remove(elemento)
        # self = list(set(self))

    def mediana(self):
        tamano = len(self)
        for elemento in self:
            if not isinstance(elemento, int):
                return
        aux = sorted(self)
        if not tamano % 2:
            return (aux[int(tamano / 2)] + aux[int(tamano / 2 - 1)]) / 2
        else:
            return aux[int(tamano // 2)]

    def aleatorizar(self):
        aux = self.copy()
        for _ in self:
            self.pop(0)
            self.append(aux.pop(randint(0, len(aux) - 1)))

    def moda(self):
        maxcount = 0
        moda = None
        for i in self:
            n = self.count(i)
            if n > maxcount:
                maxcount = n
                moda = i
        return moda

    def eliminar_pares(self):
        pares = []
        for elemento in self:
            if isinstance(elemento, int) and not elemento % 2:
                pares.append(elemento)
        for elemento in pares:
            self.remove(elemento)

    def eliminar_impares(self):
        impares = []
        for elemento in self:
            if isinstance(elemento, int) and elemento % 2:
                impares.append(elemento)
        for elemento in impares:
            self.remove(elemento)


lista = Lista2(["hola", "mundo", "mundo", 1, 1, 1, 2, 3, 7])
print(f"Lista original: {lista}")
print(f"Moda: {lista.moda()}")
lista.repetidos()
print(f"Lista sin repetidos: {lista}")
print()

lista = Lista2([1, 2, 3, 4, 5, 6, 7, 8, 9])
print(f"Lista original: {lista}")
lista.aleatorizar()
print(f"Lista aleatorizada: {lista}")
print(f"Mediana: {lista.mediana()}")
```
